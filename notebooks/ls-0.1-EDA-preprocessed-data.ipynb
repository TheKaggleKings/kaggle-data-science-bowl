```python
extract_modified['changed_first']=extract_modified['title'] != extract_modified['title'].shift(1).fillna(extract_modified['title'])
extract_modified['changed_last']=extract_modified['title'] != extract_modified['title'].shift(-1).fillna(extract_modified['title'])
```

```python
extract_modified.head(5)
```

```python
extract_modified['changed_cum']=extract_modified['changed_first'] | extract_modified['changed_last']
```

```python
extract_modified
```

```python
extract_sorted=extract_modified[(extract_modified['changed_cum']==True)][['title','timestamp']]
extract_sorted
```

```python
extract_sorted=extract_modified[(extract_modified['changed_cum']==True) & (extract_modified['type']!='Clip')][['title','timestamp']]
extract_sorted
```

```python
extract_sorted['delta']=extract_sorted['timestamp'].diff()
extract_sorted
```
