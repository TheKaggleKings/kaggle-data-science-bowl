```python
import pandas as pd
import sys
```

```python
test = pd.read_csv(
    '../data/raw/test.csv', 
    parse_dates = ['timestamp'],
)
print(test.dtypes)
print(sys.getsizeof(test) / 1024 / 1024)
print(test.memory_usage() / 1024 / 1024)
test
```

```python
df = test.copy()
df = df.astype({'world': 'category', 'type': 'category', 'title': 'category', 'event_code': 'category', 'installation_id': 'category', 'game_session': 'category', 'event_id': 'category'})
df.loc[:, 'game_time'] = pd.to_numeric(df.game_time, downcast = 'integer')
df.loc[:, 'event_count'] = pd.to_numeric(df.event_count, downcast = 'integer')
print(sys.getsizeof(df) / 1024 / 1024)
print(df.memory_usage() / 1024 / 1024)
df
```
