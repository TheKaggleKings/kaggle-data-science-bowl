```python
def create_feature_dataframe():
    columns=['installation_id']+list((data.title.unique()))
    prodf = pd.DataFrame(columns = columns)
    return prodf
```

```python
def create_individual_feature_dataframe(installation_id_temp):
    extract2=data[data['installation_id']==installation_id_temp]
    temp=pd.DataFrame(extract2.title.unique())
    temp=temp.set_index(temp[0])
    temp[0]=1
    temp=temp.T
    temp['installation_id']=installation_id_temp
    return temp
```

```python
i=0
prodf = create_feature_dataframe()
installation_id_temp_array=list(data.installation_id.unique())
for installation_id_temp in installation_id_temp_array:
    if i%2==0:
        print(i)
    i+=1
    returned_features=create_individual_feature_dataframe(installation_id_temp)
    prodf=prodf.append(returned_features)
prodf
```
